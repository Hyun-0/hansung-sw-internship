```python
from llama_index.core import SimpleDirectoryReader
from llama_index.core import VectorStoreIndex
from llama_index.core import StorageContext
from llama_index.core import Settings
from llama_index.vector_stores.qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
from config import DATASET

from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from config import FAST_EMBED_MODEL_NAME

embed_model = HuggingFaceEmbedding(model_name=FAST_EMBED_MODEL_NAME, device='cuda')
Settings.embed_model = embed_model
Settings.chunk_size = 512

# load documents
documents = SimpleDirectoryReader(DATASET).load_data()

# qdrant client 연결
client = QdrantClient(host="43.202.33.57", port=6333)

# vector store
vector_store = QdrantVectorStore(client=client, collection_name="test1")

# storage context
storage_context = StorageContext.from_defaults(vector_store=vector_store)

index = VectorStoreIndex.from_documents(documents, storage_context = storage_context)
```
